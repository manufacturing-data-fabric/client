```python
# wir ben√∂tigen wieder ein paar python imports...
import os
import sys
from os import environ

sys.path.append(os.path.abspath(".."))

from connector_client import ConnectorClient
from sparql_builder import *  # todo: remove later
from sparql_builder import SPARQLBuilder

client = ConnectorClient(bootstrap_servers=[environ.get("KAFKA_BROKER")])
builder = SPARQLBuilder()

await client.load_connector_config()
await client.return_connectors()
```

```python
query = builder.build_list_instances_query("df:Device")

output = await client.query_graphdb(query)
output
```

```python
device = f"http://stephantrattnig.org/instances#{output.loc[0]["instance"]}"

query = builder.build_get_related_query(
    subject_uri=device,
    predicate_uri="df:hasDataPoint",
)
output = await client.query_graphdb(query)
output
```

```python
datapoint_uri = f"http://stephantrattnig.org/instances#{output.loc[0]["object"]}"

query = builder.build_get_properties_query(
    subject_uri=datapoint_uri
)

output = await client.query_graphdb(query)
output
```
